**Line chart: Monthly Sales Data**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset: Monthly Sales Data
monthly_sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June',
              'July', 'August', 'September', 'October', 'November', 'December'],
    'Sales (Peso)': [5000, 7000, 8000, 6000, 9000, 12000,
                     15000, 13000, 11000, 14000, 16000, 18000]
}

# Convert the data into a DataFrame
sales_df = pd.DataFrame(monthly_sales_data)

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(sales_df['Month'], sales_df['Sales (Peso)'], marker='o', linestyle='-', color='black')

# Set title and labels
plt.title('Monthly Sales Data (Line Chart)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Sales (Peso)', fontsize=14)
plt.xticks(rotation=45)  # Rotate month labels for better visibility
plt.grid()

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

**Bar chart: Product Sales Data**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset: Product Sales Data
product_sales_data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E', 'Product F'],
    'Units Sold': [150, 300, 200, 400, 250, 350]
}

# Convert the data into a DataFrame
sales_df = pd.DataFrame(product_sales_data)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(sales_df['Product'], sales_df['Units Sold'], color='pink')

# Set title and labels
plt.title('Product Sales Data', fontsize=16)
plt.xlabel('Product Names', fontsize=14)
plt.ylabel('Number of Units Sold', fontsize=14)
plt.xticks(rotation=45)  # Rotate product labels for better visibility
plt.grid(axis='y')  # Add grid lines for y-axis

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

**Bubble plot: Product Sales and Market Share**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset: Product Sales and Market Share
product_sales_data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'Sales ($)': [50000, 75000, 60000, 90000, 30000],
    'Market Share (%)': [15, 25, 20, 30, 10],
    'Profit ($)': [15000, 20000, 18000, 30000, 5000]
}

# Convert the data into a DataFrame
sales_df = pd.DataFrame(product_sales_data)

# Plotting the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(sales_df['Sales ($)'], sales_df['Market Share (%)'],
            s=sales_df['Profit ($)'] / 100,  # Adjust size for better visibility
            alpha=0.5, color='red', edgecolors='w', linewidth=2)  # Set bubble color to pink

# Set labels and title
plt.title('Product Sales and Market Share Bubble Plot', fontsize=16)
plt.xlabel('Sales ($)', fontsize=14)
plt.ylabel('Market Share (%)', fontsize=14)

# Annotating each product on the plot
for i in range(sales_df.shape[0]):
    plt.text(sales_df['Sales ($)'][i], sales_df['Market Share (%)'][i],
             sales_df['Product'][i], fontsize=9, ha='center', va='center')

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.grid()
plt.show()
```

**Area plot and stacked plot - Monthly Revenue by Product Category**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset: Monthly Revenue by Product Category
monthly_revenue_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June',
              'July', 'August', 'September', 'October', 'November', 'December'],
    'Product A': [3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000, 8500],
    'Product B': [2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500],
    'Product C': [1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000],
    'Product D': [1000, 1200, 1500, 1800, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500]
}

# Convert the data into a DataFrame
revenue_df = pd.DataFrame(monthly_revenue_data)

# Set the Month as the index
revenue_df.set_index('Month', inplace=True)

# Define colors for the area plot
area_colors = ['#ff9999', '#66b3ff', '#99ff99']

# Plotting the Area Plot
plt.figure(figsize=(10, 6))
revenue_df.plot.area(alpha=0.5, color=area_colors)
plt.title('Monthly Revenue by Product Category', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Revenue ($)', fontsize=14)
plt.xticks(rotation=45)  # Rotate month labels for better visibility
plt.grid()
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()

# Define colors for the stacked bar plot
bar_colors = ['#ffcc99', '#c2c2f0', '#ffb3e6']

# Plotting the Stacked Plot
plt.figure(figsize=(10, 6))
revenue_df.plot(kind='bar', stacked=True, color=bar_colors)
plt.title('Monthly Revenue by Product Category', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Revenue ($)', fontsize=14)
plt.xticks(rotation=45)  # Rotate month labels for better visibility
plt.grid(axis='y')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

**Pie chart - Market Share by Product Category**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset: Market Share by Product Category
market_share_data = {
    'Product Category': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'Market Share (%)': [25, 35, 20, 15, 5]
}

# Convert the data into a DataFrame
market_share_df = pd.DataFrame(market_share_data)

# Define a list of colors
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(market_share_df['Market Share (%)'],
        labels=market_share_df['Product Category'],
        autopct='%1.1f%%',  # Display percentage in the slices
        startangle=140,     # Start angle for the first slice
        colors=colors)      # Use the defined colors

# Set title
plt.title('Market Share by Product Category', fontsize=16)

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

**Polar chart: Product Performance Metrics**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame for the product performance metrics
data = {
    'Metric': ['Quality', 'Durability', 'Price', 'Customer Service', 'Design', 'Features'],
    'Product A': [8, 9, 7, 6, 8, 9],
    'Product B': [7, 8, 6, 9, 7, 8],
    'Product C': [9, 7, 8, 8, 9, 6],
    'Product D': [6, 5, 9, 7, 8, 7],
}

df = pd.DataFrame(data)

# Number of metrics
num_metrics = len(df['Metric'])

# Create a list of angles for the metrics
angles = np.linspace(0, 2 * np.pi, num_metrics, endpoint=False).tolist()

# The plot is circular, so we need to "complete the loop" and append the start value to the end
values = df.loc[:, 'Product A'].tolist() + [df.loc[0, 'Product A']]
angles += angles[:1]

# Initialize the polar plot
plt.figure(figsize=(8, 8), dpi=80)
ax = plt.subplot(111, polar=True)

# Define colors for each product
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Plot each product
for i, product in enumerate(df.columns[1:]):
    values = df[product].tolist() + [df.loc[0, product]]
    ax.fill(angles, values, color=colors[i], alpha=0.25)  # Use color from the list
    ax.plot(angles, values, color=colors[i], label=product)  # Use color from the list

# Add labels for each metric
ax.set_xticks(angles[:-1])
ax.set_xticklabels(df['Metric'])

# Add a title and legend
plt.title('Product Performance Metrics (Polar Chart)', size=15, color='black')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

# Show the plot
plt.show()
```

**Lollipop chart: Product Sales Figures**

```python
import matplotlib.pyplot as plt

# Data
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E', 'Product F']
sales = [5000, 8000, 6500, 12000, 7000, 9000]

# Create a lollipop chart
plt.figure(figsize=(10, 6))
plt.stem(products, sales, markerfmt='o', linefmt='purple', basefmt=' ', use_line_collection=True)

# Adding titles and labels
plt.title('Product Sales Figures', fontsize=16)
plt.xlabel('Products', fontsize=14)
plt.ylabel('Sales (Peso)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the chart
plt.tight_layout()
plt.show()
```
