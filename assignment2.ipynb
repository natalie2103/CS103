**Nominal Scale**

This is the most basic measurement scale. It involves categorizing data without any order or ranking. The numbers or labels in this scale are used only to identify categories or groups.

**Barplots**

The bars represent the number of individuals who prefer each type of fruit. The fruits included in the graph are apple, banana, orange, and grape.
Apple gained a frequency of 2.00, and the most popular fruit among the respondents. Meanwhile, banana, orange, and grape gained around 1.00 frequency, indicating a relatively equal preference among the respondents.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Sample dataset
data = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Favorite Fruit': ['Apple', 'Banana', 'Orange', 'Apple', 'Grape']
})

#Function to create a bar plot for nominal data
def plot_barplot_nominal(data, column):
    #Count the frequency of each category
    value_counts = data[column].value_counts().reset_index()
    value_counts.columns = [column, 'Count']

    plt.figure(figsize=(8, 6))  #Set the size of the plot
    sns.barplot(x=column, y='Count', data=value_counts, palette='BuGn')
    plt.title('Favorite Fruit')
    plt.xlabel(column)
    plt.ylabel('Frequency')

    plt.show()

#Plotting the bar chart for Favorite Fruit
plot_barplot_nominal(data, 'Favorite Fruit')
```

**Ordinal Scale**

Involves data that can be categorized and ranked. However, the differences between the ranks are not necessarily equal.

**Pie Chart**

The pie chart presents the customer satisfaction data, categorized into three: Satisfied, Not Satisfied, and Very Satisfied. Not Satisfied gained 38.3% of the total responses resulting to highest percentage. Satisfied is the second most popular category with 35.0% of the total responses. Representing 26.7% of the total responses, Very Satisfied is the least popular category.


```python
import matplotlib.pyplot as plt
import pandas as pd

#Assuming we have 3 responses from customers on 3 different aspects
customer_data = pd.DataFrame({
    'Satisfied': [4, 5, 3, 4, 5],
    'Very Satisfied': [2, 3, 4, 3, 4],
    'Not Satisfied': [5, 4, 5, 4, 5]
})

#Calculate the average of each response
average_responses = customer_data.mean()

#Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(average_responses, labels=average_responses.index, autopct='%1.1f%%', startangle=90)
plt.title('Customer Satisfaction')
plt.show()
```

**Line Chart**

The line chart shows a varying pattern of temperature readings. There are both increases and decreases in temperature throughout the data points. The highest temperature recorded is approximately 30 degree celsius while the lowest temperature recorded is approximately 22 degree celsius

```python
import plotly.graph_objs as go
import pandas as pd
!pip install plotly
import plotly.io as pio

#Sample dataset based on the image (ID and Temperature)
data = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Temperature (°C)': [25, 28, 22, 30, 27]
})

#Creating the trace (line graph)
trace = go.Scatter(
    x = data['ID'],
    y = data['Temperature (°C)'],
    mode = "lines+markers",  #Line graph with markers on points
    name = "Temperature",
    marker = dict(color = 'rgba(16, 112, 2, 0.8)')  #Marker color
)

#Defining the layout
layout = dict(
    title = dict(text='Temperature Readings', x=0.5), #Center the title using x=0.5
    xaxis = dict(title= 'ID', ticklen= 5, zeroline= False),
    yaxis = dict(title= 'Temperature (°C)'),
)

#Combine data and layout into a figure
fig = dict(data=[trace], layout=layout)

#Plot the figure using pio.show()
pio.show(fig)
```

**Scatter Plot**

The scatter plot visually represents the relationship between height and weight for a given set of data points. As height increases, weight tends to increase as well. Example, taller individuals tend to weigh more than shorter individuals.



```python
import matplotlib.pyplot as plt

#Sample dataset: Height and Weight
data = {'ID': [1, 2, 3, 4, 5],
        'Height (cm)': [170, 165, 180, 175, 160],
        'Weight (kg)': [70, 65, 80, 75, 60]}

#Convert to DataFrame
df = pd.DataFrame(data)

#Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Height (cm)'], df['Weight (kg)'], color='pink')

#Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')

#Show the plot
plt.show()
```
